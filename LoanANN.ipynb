```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
```

```python
df=pd.read_csv("CleanedLoanData.csv")
```

```python
df.head()
```

```python
x=df.drop("loan_status",axis=1)
y=df["loan_status"]
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=77)
```

```python
from keras.models import Sequential
ann=Sequential()
from keras.layers import Dense
from keras.layers import Dropout
```

```python
ann.add(Dense(input_dim=19,units=38,activation="relu",kernel_initializer="uniform"))
# ann.add(Dense(units=38,activation="relu",kernel_initializer="uniform"))
ann.add(Dense(units=1,activation="sigmoid",kernel_initializer="uniform"))
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=30)
```

```python
train_pry=ann.predict(x_train)
train_pry=train_pry>0.5
from sklearn.metrics import accuracy_score
TrainAccuracy=accuracy_score(y_train,train_pry)
TrainAccuracy
```

```python
test_pred=ann.predict(x_test)
test_pred=test_pred>0.5
```

```python
TestAccuracy=accuracy_score(y_test,test_pred)
TestAccuracy
```

```python
def cv():
    from keras.models import Sequential
    ann=Sequential()
    ann.add(Dense(input_dim=19,units=38,activation="relu",kernel_initializer="uniform"))
    # ann.add(Dense(units=38,activation="relu",kernel_initializer="uniform"))
    ann.add(Dense(units=1,activation="sigmoid",kernel_initializer="uniform"))
    ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
    return ann
```

```python
from scikeras.wrappers import KerasClassifier
model=KerasClassifier(cv(),batch_size=32,epochs=30)
from sklearn.model_selection import cross_val_score
cv=cross_val_score(model,x_train,y_train,cv=5,scoring="accuracy").mean()
cv
```
